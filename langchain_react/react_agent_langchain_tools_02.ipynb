```python
def get_react_agent(llm, tools, prompt, max_iterations=5):
    # Construct the ReAct agent
    agent = create_react_agent(llm, tools, prompt)
    # Create an agent executor by passing in the agent and tools
    agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True, max_iterations=max_iterations)

    return agent_executor
```

## **RAG como tool**

[RAG (Retrieval Augmented Generation)](https://www.promptingguide.ai/research/rag) es un patron para LLMs que permite ajustar la respuesta de un LLM para que en vez de ser "generica" se base (o enfoque) en informacion provista por el usuario. Normalmente esta informacion se encuentra almacenada en una base vectorial.

En nuestro caso, vamos a tomar un archivo de texto como fuente de informacion, y vamos a aplicarle embeddings.

```python
!wget "https://raw.githubusercontent.com/V-Sher/LangChain_ReAct_Demo/refs/heads/main/data/globalcorp_hr_policy.txt"
```

```python
# Get the embeddings. Alternatively, we can use OpenAI embeddings
# embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
embeddings = OpenAIEmbeddings(model="text-embedding-3-small", api_key=userdata.get('OPENAI_API_KEY'))

# Test embeddings
text = "This is a test document."
query_result = embeddings.embed_query(text)
print(len(query_result))
query_result[:5]
```

```python
# Get the text as a set of documents
loader = TextLoader("/content/globalcorp_hr_policy.txt")
documents = loader.load()
print(len(documents), "documents")
```

```python
display_markdown(documents[0].page_content, raw=True) # Print the text (document)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splits = text_splitter.split_documents(documents)
print(len(splits), "splits")

# For simplicity, let's use a Chroma database in memory
vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)

# We create a retriever for the database
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
# Define a RAG chain (for testing)
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
)
```

```python
question3 = "What are the communication initiatives?"
# question3 = "What is the allocated budget for communication initiatives?"

qa_chain.invoke(question3)
```

```python
# Convert the retriever into a tool
retriever_tool = create_retriever_tool(
    retriever,
    "search_hr_policy",
    "Searches and returns excerpts from the HR policy.",
)
```

## **Tool para conversiones de moneda**

```python
# Test the library
message = "I want 2 liters of wine and 5 million $"
quantities = parser.parse(message)
quantities # [0].value
```

```python
# Test the library
c = CurrencyConverter()
# amount = 50000000
amount = quantities[1].value
print(amount)
c.convert(amount, 'JPY', 'USD')
```

```python
converter = CurrencyConverter()
def convert_currency_to_usd(amount: str) -> int:
    "Converts the currency amount from any country into USD."
    # print(amount)

    if "¥" in amount:
        exclude_symbol = amount.replace("¥", "")
        amount = parser.parse(exclude_symbol)[0].value
        return converter.convert(amount, "JPY", "USD")
    if "$" in amount:
        exclude_symbol = amount.replace("$", "")
        amount = parser.parse(exclude_symbol)[0].value
        return amount
    if "€" in amount:
        exclude_symbol = amount.replace("€", "")
        amount = parser.parse(exclude_symbol)[0].value
        return converter.convert(amount, "EUR", "USD")
    if "JPY" in amount:
        amount = parser.parse(amount)[0].value
        return converter.convert(amount, "EUR", "USD")
    if "EUR" in amount:
        amount = parser.parse(amount)[0].value
        return converter.convert(amount, "JPY", "USD")
    if "USD" in amount:
        amount = parser.parse(amount)[0].value
        return amount

# It is okay to define single-input tools in this manner.
currency_conversion_tool = Tool(
    name="currency_conversion",
    func=convert_currency_to_usd,
    description="Useful for converting any currency into USD. Input should be a currency amount.",
)
```

## **Tool para cálculo de diferencia (porcentajes)**

```python
# For multi-input tool, useful to define schema class
class Metrics(BaseModel):
    num1: float = Field(description="The first value")
    num2: float = Field(description="The second value")


@tool(args_schema=Metrics)
def percentage_difference(num1: float, num2:float) -> float:
    """Calculates the percentage difference between two numbers"""
    num1 = float(num1)
    num2 = float(num2)
    return (np.abs(num1 - num2) / ((num1 + num2) / 2)) * 100

@tool
def percentage_difference_tool(input: Union[str, Metrics]) -> float:
    """Calculates the percentage difference between two numbers."""
    if isinstance(input, str):
        try:
            # Attempt to parse the input string as two comma-separated numbers
            num1, num2 = map(float, input.split(","))
        except ValueError:
            return "Error: Invalid input format. Please provide two comma-separated numbers."
    elif isinstance(input, Metrics):
        num1 = input.num1
        num2 = input.num2
    else:
        return "Error: Invalid input type. Input must be a string or a Metrics object."

    # Calculate percentage difference
    if num2 == 0:
        return "Error: Cannot calculate percentage difference when the second number is 0."

    diff = (num1 - num2) / num2 * 100
    return diff
    # return f"The percentage difference between {num1} and {num2} is {diff:.2f}%."
```
